```python
from bs4 import BeautifulSoup
import requests
from random import randint
from time import sleep
import pandas as pd
```

```python
response = requests.get('https://www.emsc-csem.org/Earthquake/')
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
response.content
```

```python
date_tags = soup.select('.tbdat')
lat_tags = soup.select('.tblat')
lon_tags = soup.select('.tblon')
mag_tags = soup.select('.tbmag')
reg_tags = soup.select('.tbreg')

lat_tags
```

```python
date = [tag.text for tag in date_tags]
lat = [tag.text for tag in lat_tags]
lon = [tag.text for tag in lon_tags]
mag = [tag.text for tag in mag_tags]
reg = [tag.text for tag in reg_tags]
```

```python
df = pd.DataFrame({'Date/Time': date, 'Lat': lat, 'Lon':lon, 'Magnitude': mag, 'Region': reg})
df
```

```python
import websocket
import time
import json

all_data = []
def on_message(ws, message):
    print(f"Received message: {message}")
    # Convert the received message from JSON string to a Python dictionary
    message_data = json.loads(message)
    
    # Append the 'data' part of the message to all_data
    all_data.append(message_data['data'])

def on_error(ws, error):
    print(f"Error: {error}")

def on_close(ws, close_status_code, close_msg):
    print("Connection closed")
    # Convert all_data into a Pandas DataFrame when the connection is closed
    df = pd.DataFrame(all_data)
    print(df)

def on_open(ws):
    print("Connection opened")
    # If you need to send any initial message upon connection:
    # ws.send(json.dumps({"type": "YOUR_TYPE_HERE", "data": "YOUR_DATA_HERE"}))
    # Close the connection after 15 minutes (900 seconds)
    threading.Timer(900, close_connection, args=(ws,)).start()

# Create a new WebSocket connection
ws = websocket.WebSocketApp("wss://cobra.emsc-csem.org/eq_search",
                            on_message=on_message,
                            on_error=on_error,
                            on_close=on_close)

ws.on_open = on_open
ws.run_forever()
```

```python
earthquake_data = []

for feature in all_data[1:]:
    properties = feature['properties']
    time_str = properties['time']['time_str']
    lat = properties['location']['lat']
    lon = properties['location']['lon']
    mag = properties['magnitude']['mag']
    region = properties['place']['region']

    earthquake_data.append([time_str, lat, lon, mag, region])

# Convert to DataFrame
df_earthquake = pd.DataFrame(earthquake_data, columns=['Date and Time', 'Latitude', 'Longitude', 'Magnitude', 'Region'])

df_earthquake.drop_duplicates().reset_index(drop=True)
```
